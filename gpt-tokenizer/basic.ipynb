```python
def print_dict(x):
    print(sorted(((v, k) for k,v in x.items()), reverse=True))
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

stats = get_stats(tokens)
print_dict(stats)
```

```python
def get_next_merge(corpus):
    stats = get_stats(corpus)
    best_pair = max(stats, key=stats.get)
    return best_pair

def merge(corpus, pair, new_id):
    new_corpus = []
    i = 0 
    while i < len(corpus):
        if (
            i < len(corpus) - 1 and 
            corpus[i]==pair[0] and corpus[i+1]==pair[1]
        ):
            new_corpus.append(new_id)
            i += 2
        else:
            new_corpus.append(corpus[i])
            i += 1
    return new_corpus

print(merge([5, 6, 6, 7, 9, 1], (6, 7), 99))
```

```python
# making the training text longer to have more representative token statistics
# text from https://www.reedbeta.com/blog/programmers-intro-to-unicode/
with open("training.txt") as f:
    text = f.read()

tokens = text.encode("utf-8") # raw bytes
tokens = list(map(int, tokens)) # convert to a list of integers in range 0..255 for convenience
print(len(tokens))
```

```python
vocab_size = 276 # the desired final vocabulary size
num_merges = vocab_size - 256
ids = list(tokens) # copy so we don't destroy the original list

merges = {} # (int, int) -> int
for i in range(num_merges):
  pair = get_next_merge(ids)
  idx = 256 + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx
```

```python
print("tokens length:", len(tokens))
print("ids length:", len(ids))
print(f"compression ratio: {len(tokens)/len(ids):.2f}")
```

```python

unmerge = {v: k for k,v in merges.items()}
def bytes_of_token(token: int) -> List[int]:
    if token < 256:
        return [token]
    else:
        first,second = unmerge[token]
        return bytes_of_token(first) + bytes_of_token(second)

vocab = [bytes_of_token(i) for i in range(vocab_size)]
print(vocab)

def decode(ids):
    byte_groups = [vocab[token] for token in ids]
    encoded = bytes([x for group in byte_groups for x in group])
    return encoded.decode('utf-8', errors='replace')
```

```python
print(decode([101, 32, 101, 32]))
print(decode([128]))
```

```python
inverse_vocab = {tuple(x): i for i,x in enumerate(vocab)}
longest_token = max([len(x) for x in inverse_vocab])
print(longest_token)
print(inverse_vocab)
def incorrect_but_tempting_encode(text):
    ids = []
    remaining = list(text.encode('utf-8'))

    while remaining:
        lookahead = longest_token
        while tuple(remaining[:lookahead]) not in inverse_vocab:
            lookahead -= 1
            assert lookahead > 0
        
        ids.append(inverse_vocab[tuple(remaining[:lookahead])])
        remaining = remaining[lookahead:]
    
    return ids
```

```python
def encode(text):
  # given a string, return list of integers (the tokens)
  tokens = list(text.encode("utf-8"))
  while len(tokens) >= 2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float("inf")))
    if pair not in merges:
      break # nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)
  return tokens
```

```python
print(decode(encode("hello world")))
text2 = decode(encode(text))
print(text2 == text)

valtext = "Many common characters, including numerals, punctuation, and other symbols, are unified within the standard and are not treated as specific to any given writing system. Unicode encodes thousands of emoji, with the continued development thereof conducted by the Consortium as a part of the standard.[4] Moreover, the widespread adoption of Unicode was in large part responsible for the initial popularization of emoji outside of Japan. Unicode is ultimately capable of encoding more than 1.1 million characters."
print(decode(encode(valtext))==valtext)

print(incorrect_but_tempting_encode(valtext)==encode(valtext))
```
